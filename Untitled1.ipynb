```python
from bokeh.plotting import figure, output_notebook, show
from pandas import pandas as pd
from bokeh.palettes import Spectral11
```

```python
df = pd.read_csv('gammon_ham.csv',  names=['time','pit', 'food1', 'fan', 'food2'], header=0, index_col=0)
df['pit'] = df.apply(lambda row: ((float(row['pit']/10) - 32) / 1.8),axis=1)
df['food1'] = df.apply(lambda row: ((float(row['food1']/10) - 32) / 1.8),axis=1)
#df['food2'] = df.apply(lambda row: ((float(row['food2']/10) - 32) / 1.8),axis=1)
df['target'] = 65
df['pit target'] = 180

del df['food2']
numlines=len(df.columns)
mypalette=Spectral11[0:numlines]

p = figure(width=500, height=300)#, x_axis_type="datetime") 
p.multi_line(xs=[df.index.values]*numlines,
                ys=[df[name].values for name in df],
                line_color=mypalette,
                line_width=2)

output_notebook()


show(p)
```
